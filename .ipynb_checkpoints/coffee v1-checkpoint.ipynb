```python
# Make a bar plot for types of coffee, sort first
# Calculate value counts and sort them
category_counts = df["coffee_name"].value_counts().sort_values(ascending=False)

# Get the sorted order of categories
sorted_categories = category_counts.index.tolist()

# Convert the column to a Categorical type with the desired order
df["coffee_name"] = pd.Categorical(df["coffee_name"], categories=sorted_categories, ordered=True)

# Make the plot
sns.histplot(data = df, x = "coffee_name")
plt.xticks(rotation=45)  # rotate 45 degrees
plt.show()
```

Americano with milk, Latte, Americano w/o milk are the top 3 types of coffee sales (count). 

```python
# Make a histogram for hour of the day purchases are made

sns.histplot(data = df, x = "hour_of_day")
plt.xticks(rotation=45) 
plt.show()
```

A bit of a surge in the late AM and late afternoon for coffee sales. 

```python
# Make a histogram for day of the week purchases happen

# Set order of categories
order = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
        
# Make plot
sns.countplot(data = df, x = "Weekday", order=order)
plt.xticks(rotation=45)  # rotate 45 degrees
plt.show()
```

Early in the week and again on Friday slight spike in sales

```python
# Make a histogram for month that purchases occur

order2 = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

sns.countplot(data = df, x = "Month_name", order=order2)
plt.xticks(rotation=45) 
plt.show()
```

Feb, Mar, Oct top 3 months in sales. 

```python
# Also want to look at some correlations, beginning with type of beverage and money from total sales

# perform aggregations
df_agg1 = df.groupby("coffee_name", as_index=False)["money"].sum()

# sort by sum (descending)
df_agg1 = df_agg1.sort_values("money", ascending=False)

# do plot
sns.barplot(data = df_agg1, x = "coffee_name", y = "money")
plt.xticks(rotation=45)
plt.show()
```

Latte edges out Americano with Milk for total sales, Cappucino number three. 

```python
# Calculate and plot average cost per sale for each type of beverage

# perform aggregations
df_agg2 = df.groupby("coffee_name", as_index=False)["money"].mean()

# sort by sum (descending)
df_agg2 = df_agg2.sort_values("money", ascending=False)

# make plot
ax = sns.barplot(df_agg2, x = "coffee_name", y = "money")
ax.bar_label(ax.containers[0], fmt="%.1f")  # containers[0] = the bars in the plot
plt.xticks(rotation=45)
plt.show()
```

Now looking at mean sales instead of sum sales. Cappucino and Latte cost more than Americano options, accounting for them having higher total revenue even though they sold less. Latte and Americanos were top three for count of sales. 

```python
# Just need to sort these values by the y axis
# Create categories list from dataframe "coffee_name" column
categories = (
    df.groupby("coffee_name")["money"]
      .agg(count="count", total_sales="sum")
      .reset_index()
)

# Place data in format to be used in the plots
categories_count = categories.sort_values(by=["count"], ascending=False)
categories_sales = categories.sort_values(by=["total_sales"], ascending=False)

# Explicitly order due to issues in getting Seaborn to order
order_count = categories_count["coffee_name"]
order_sales = categories_sales["coffee_name"]

# Do a side-by-side axis in figure
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Left subplot
sns.barplot(x="coffee_name", y="count", data=categories, order=order_count, ax=axes[0])
axes[0].set_title("Total Count of Sales")

# Right subplot
sns.barplot(x="coffee_name", y="total_sales", data=categories, order=order_sales, ax=axes[1])
axes[1].set_title("Total Amount of Sales ($)")

# Plot
for ax in axes:
    ax.tick_params(axis="x", rotation=45)
plt.tight_layout()
plt.show()
```

Here we can see the difference made between Americano with and without milk (high volume) but latte and cappucino "jump" these when you're talking total sales. 

Where to go from here... how to plot the pricing dynamics across the course of the year for the top four sellers. This can help inform some analysis about the possibility of price changes (expected value) or best times to do some sales.

```python
# Plot dates of the year scatterplot with costs per cup

ax = sns.scatterplot(data=df, x="datetime", y="money", hue="coffee_name")

ax.legend(
    bbox_to_anchor=(1.05, 0.5),   # (x, y): 1.05 pushes right, 0.5 is vertical center
    loc="center left",            # anchor point on legend box
    borderaxespad=0               # removes extra padding
)
plt.show()
```

```python
# Now show by individual types of coffee, how their costs change
df_filter1 = df[df["coffee_name"].isin(["Americano with Milk"])]
df_filter2 = df[df["coffee_name"].isin(["Americano"])]
df_filter3 = df[df["coffee_name"].isin(["Latte"])]
df_filter4 = df[df["coffee_name"].isin(["Cappuccino"])]

# New attempt to do a side-by-side axis in figure
fig, axes = plt.subplots(4, 1, figsize=(12, 10))

# Subplot 1
sns.scatterplot(data=df_filter1, x="datetime", y="money", hue="coffee_name", ax=axes[0], legend=False)
axes[0].set_title("Price of Americano type with Milk")

# Subplot 2
sns.scatterplot(data=df_filter2, x="datetime", y="money", hue="coffee_name", ax=axes[1])
axes[1].set_title("Price of Americano")

# Subplot 3
sns.scatterplot(data=df_filter3, x="datetime", y="money", hue="coffee_name", ax=axes[2], legend=False)
axes[2].set_title("Price of Latte")

# Subplot 4
sns.scatterplot(data=df_filter4, x="datetime", y="money", hue="coffee_name", ax=axes[3], legend=False)
axes[3].set_title("Price of Cappuccino")

axes[1].legend(
    bbox_to_anchor=(1.05, 0.5),   # (x, y): 1.05 pushes right, 0.5 is vertical center
    loc="center left",            # anchor point on legend box
    borderaxespad=0               # removes extra padding
)

for ax in axes.ravel():
    ax.set_ylim(20, 45)  

plt.tight_layout()
plt.show()
```

```python
# Build 2x2 panels chart showing day of week sales count for each top 4 type of coffee; idea of variance for each, calculate stats

# Start with filters for each
df_filter5 = df[df["coffee_name"].isin(["Americano with Milk"])]
df_filter6 = df[df["coffee_name"].isin(["Americano"])]
df_filter7 = df[df["coffee_name"].isin(["Latte"])]
df_filter8 = df[df["coffee_name"].isin(["Cappuccino"])]

# Set up the figure of four charts
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

# Set order of categories and calculate max/min values
order = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

# Now do each chart
sns.countplot(data = df_filter5, x = "Weekday", order=order, ax=axes[0, 0])
axes[0, 0].set_title("Count of Sales, Americano type with Milk")
mean = df_filter5["Weekday"].value_counts().mean()
axes[0, 0].axhline(mean, ls="--", color="blue")
axes[0, 0].annotate(f"Mean = {mean:.1f}",
                    xy = (3, mean), 
                    xytext = (1, 5), textcoords="offset points", 
                    fontsize=10, ha="center", va="bottom")

sns.countplot(data = df_filter6, x = "Weekday", order=order, ax=axes[0, 1])
axes[0, 1].set_title("Count of Sales, Americano")
mean = df_filter6["Weekday"].value_counts().mean()
axes[0, 1].axhline(mean, ls="--", color="blue")
axes[0, 1].annotate(f"Mean = {mean:.1f}",
                    xy = (5.5, mean), 
                    xytext = (1, 5), textcoords="offset points", 
                    fontsize=10, ha="center", va="bottom")

sns.countplot(data = df_filter7, x = "Weekday", order=order, ax=axes[1, 0])
axes[1, 0].set_title("Count of Sales, Latte")
mean = df_filter7["Weekday"].value_counts().mean()
axes[1, 0].axhline(mean, ls="--", color="blue")
axes[1, 0].annotate(f"Mean = {mean:.1f}",
                    xy = (5.5, mean), 
                    xytext = (1, 5), textcoords="offset points", 
                    fontsize=10, ha="center", va="bottom")

sns.countplot(data = df_filter8, x = "Weekday", order=order, ax=axes[1, 1])
axes[1, 1].set_title("Count of Sales, Cappuccino")
mean = df_filter8["Weekday"].value_counts().mean()
axes[1, 1].axhline(mean, ls="--", color="blue")
axes[1, 1].annotate(f"Mean = {mean:.1f}",
                    xy = (5, mean), 
                    xytext = (1, 5), textcoords="offset points", 
                    fontsize=10, ha="center", va="bottom")

# Make plot
plt.subplots_adjust(wspace=0.3, hspace=0.3) # wspace for horizontal, hspace for vertical
plt.show()
```
